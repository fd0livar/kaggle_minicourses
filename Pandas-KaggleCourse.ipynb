### Creating data  
There are two core object in pandas: the **DataFrame** and the **Series**.

#### DataFrame  
A DataFrame is a table. It contains an array of individual entries, each of which has a certain value. Each entry corresponds to a row (or record) and a column. 

For example:

```python
pd.DataFrame({'Yes':[50, 21], 'No': [131, 2]})
```

Entries of a DataFrame can also be strings.

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 
              'Sue': ['Pretty good.', 'Bland.']})
```

We are using the `pd.DataFrame()` constructor to generate DataFrame objects. The syntax for declaring a new one is a *dictionary* whose keys are the column names, and whose values are a list of entries. This is the standard way of constructing a new DataFrame. 

The list of row labels in a DataFrame is known as an **index**. We can assign values to it by using an `index` parameter in the constructor. 

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'],
              'Sue': ['Pretty good.', 'Bland.']}, 
             index=['Product A', 'Product B'])
```

#### Series  
A Series is a sequence of data values. One can a create one as a list:

```python
pd.Series([1, 2, 3, 4, 5])
```

One can set values with an `index` parameter as with the DataFrame. However, it does not have a column name, it only has one overall `name`:

```python
pd.Series([30, 35, 40], index=['2015 Sales', '2016 Sales', '2017 Sales'],
         name='Product A')
```

Series and Dataframes are closely related.

## Indexing, Selecting & Assigning

It is important to select the relevant data point quickly and effectively.

### Naive accessors

Pandas carries over all the tools of Python for accessing data. For example:

```python
wine_reviews.country
```

In a Python dict, you access its values using the indexing `[]` operator. We can access in the same way the columns in a DataFrame:

```python
wine_reviews['country']
```

The indexing operator `[]` is better as it can handle special characters such as blank-space. 

Following the dictionary accessing, you can drill down to a specific value by using the indexinf opereator once more:

```python
wine_reviews['country'][0]
```

### Indexing in Pandas

In addition to those Python opreators, Pandas has its own operators: `loc` and `iloc`. These are used for more advanced operations:

#### Index-based selection  

This is one of the two paradigms for indexign: Selecting data based on its numerical position in data. `iloc` follows this paradigm. 

To select the first row of data in a DataFrame, we may use the following:

```python
wine_reviews.iloc[0]
```

Both `loc` and `iloc` are **row-first**, **column-second**. The opposite of what is done in native Python. 

To get a column with `iloc` we do:

```python
wine_reviews.iloc[:, 0]
```

We can combine the `:` operator to obtain a range of the whole column:

```python
wine_reviews.iloc[:3, 0]
```

```python
wine_reviews.iloc[1:3, 0]
```

It is also possible to pass a list:

```python
wine_reviews.iloc[[0, 1, 2], 0]
```

Negative numbers can also be used in the selection. Here are the last five elements of the dataset:

```python
wine_reviews.iloc[-5:]
```

#### Label-based selection

This is the second paradigm for attribute selection. This is used by the `loc` operator. The *data index value* is what matters. 

For example, to get the first entry in `wine_reviews` we do:

```python
wine_reviews.loc[0, 'country']
```

`iloc` is conceptually simpler than `loc` because it ignores the dataset indices. The first one works independently of the index of the DataFrame. The second one uses the information of the indices to work. It's usually easier to do things with `loc`. For example:

```python
wine_reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
```

#### Chossing between `loc` and `iloc`

They use different indexing schemes: `iloc` uses the Python stdlib scheme  (excluding the last element), while `loc` includes everything in the range. 

This is done because `loc` can index any type of value, a *string*, for example. Then if you want to select a range of string indices, it's more convenient to put the last object than to invent a new one to add the previous string object. *Keep this in mind!*

#### Manipulating the index

The index is very important for label-based selection. We can manipulate it, though.

The `set_index()` method can be used to do that. For example:

```python
wine_reviews.set_index('title')
```

Sometimes you can find a good index instead of the currently used bu the DataFrame.

#### Conditional selection

Here you can start to filter the data selection. 

For example, to select above-average italian wines, you can start by checking if a wine is italian or not.

```python
wine_reviews.country == 'Italy'
```

We produced a Series of `True`/`False` booleans based on the `country` of each record. This can be used inside of `loc` to select relevant data!

```python
wine_reviews.loc[wine_reviews.country == 'Italy']
```

We got less rows. Around 15% of wines originate from Italy.

Now, to get the above-average wines, we can select the wines with at least 90 points (the scale is a 80-to-100 one). 

We join the filters with the `&` symbol.

```python
wine_reviews.loc[(wine_reviews.country == 'Italy') & 
                 (wine_reviews.points >= 90)]
```

The **or** operator is the `|` (pipe):

```python
wine_reviews.loc[(wine_reviews.country == 'Italy') | 
                 (wine_reviews.points >= 90)]
```

Pandas has some built-in conditional selectors:

`isin`: this lets you select data whose value 'is in' a list of values. For example, if we want to select wines only from  Italy or France:

```python
wine_reviews.loc[wine_reviews.country.isin(['Italy', 'France'])]
```

`isnull` (and its companion `notnull`): These let you highlight values which are (or are not) empty (`NaN`). For example, to filter out wines lacking a price tag in the dataset, we do:

```python
wine_reviews.loc[wine_reviews.price.notnull()]
```

#### Assigning data

This is the opposite, assigning a value to an entry:

```python
wine_reviews['critic'] = 'everyone'
wine_reviews['critic']
```

Also an iterable of values:

```python
wine_reviews['index_backwards'] = range(len(wine_reviews), 0, -1)
wine_reviews['index_backwards']
```

## Summary Functions and Map

After selecting the data out of the DataFrame, we might have to work on it to reformat it to the way we want it.  
We'll use the Wine Magazine for demonstration.

### Summary functions

There are many pandas functions that help to restructure the data in some useful way. Here is the `describe()` method:

```python
wine_reviews.points.describe()
```

This method generates a high-level summary of the attributes of the given column. It changes according to the data type. Here is the output for string data:

```python
wine_reviews.taster_name.describe()
```

For particular simple summary statistics about a column, there are useful pandas functions to do that. For example the `mean()` function:

```python
wine_reviews.points.mean()
```

To see a list of unique values we can use the `unique()` function:

```python
wine_reviews.taster_name.unique()
```

To see a list of unique values and how often they occur in the dataset, we can use the `values_count()` method:

```python
wine_reviews.taster_name.value_counts()
```

### Maps

This term is borrowed from mathematics. This is a mapping function. Maps are useful to tranform the data to the format we want them to be in.  
There are two mapping methods that you will use often.

`map()` is the slightly simpler one. For example, if you want to *remean* the scores the wines received ot 0, we can do:

```python
review_points_mean = wine_reviews.points.mean()
wine_reviews.points.map(lambda p: p - review_points_mean)
```

The function one passes to `map()` should expect a single value from the Series and return a transformed version of that value. `map()` returns a new Series where all the values have been transformed by the function.

`apply()` is the equivalent method, but if we want to transform a whole DataFrame by calling a custom method on each row.

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

wine_reviews.apply(remean_points, axis='columns').points
```

**Note:** `map()` and `apply()` return new, transformed Series and DataFrames respectively. They don't modify the original data they're called on. 

Pandas provides common mapping operations as built-ins. For example:

```python
review_points_mean = wine_reviews.points.mean()
wine_reviews.points - review_points_mean
```

Pandas understands the context of the code and does what is expected. In the code above, it applied the mean to each row. But it also can perform operations between Series of equal length. For example, we can combine country and region if we do:

```python
wine_reviews.country + ' - ' + wine_reviews.region_1
```

These operations are faster than `map()` and `apply()` because they use speed ups built into pandas. All of the standard Python operators worn in this manner. 

However, `map()` and `apply()` can do more advanced things. They are more flexible. They can apply conditional logic, which cannot be done with addition and substraction alone.

## Grouping and Sorting

This is the next step from *Maps* as it will allow us to so something specific to a group of data, instead of a column in a DataFrame or a Series. We'll use the `groupby()` operation.

### Groupwise analysis

We can replicate the job of `value_counts()` by doing:

```python
pd.set_option('display.max_rows', 5)
```

```python
wine_reviews.groupby('points').points.count()
```

- `groupby()` creates a group of *wine_reviews* which allotted the same point values to the given wines.
- Then, for each of these groups, we grabbed the `points()` column and counted how many times it appeared. `value_counts()` is just a shorcut to this `groupby()` operation. 

We can use any of the summary functions we've used before with this data.  
For example, to get the cheapest wine in each point value category, we can do the following:


```python
wine_reviews.groupby('points').price.min()
```

Each group can be considered as a slice of our DataFrame containing only data with values that match.  
This DataFrame is accessible to us directly using the `apply()` method, and we can then manipulate the data in any way we see fit. 

For example, let's select the name of the first wine reviewed from each winery in the dataset:

```python
wine_reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

For even more fine-grained control, you can also group by more than one column. 

For example, let's pick out the best wine by country and province:

```python
wine_reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

Another important `groupby()` method is `agg()`, which lets us run a bunch of different functions on our DataFrame simultaneously.

For example, we can generate a simple statistical summary of the datasets as follows:

```python
# get the best chilean wine
# get the cheapest chilean wine
# get the best points/price chilean wine
```

```python
wine_reviews.groupby(['country']).price.agg([len, min, max])
```

A good usage of `groupby()` will allow us to do lots of powerful things with the datasets.

### Multi-indexes

Depending on the operation we run on `groupby()`, we can get different result,called a multi-index.

For example:

```python
countries_reviewed = wine_reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
type(countries_reviewed.index)
```

Multi-indices have several methods which are absent for single-level indices. *This concept is important to know it*.

There are detailed use cases with instructions on using them in the MultiIndex section in the pandas documentation. 

However, in general the multi-index method most used is the one for converting back to regular index, the `reset_index()` method:

```python
countries_reviewed.reset_index()
```

### Sorting

This is similar to the *bigquery-SQL* **SORT BY** method. Here we can sort the data by some value instead of the default value in the index.

For example:

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

The default ordering is the ascending sort. To change that we do `ascending=False`:

```python
countries_reviewed.sort_values(by='len', ascending=False)
```

To sort by index values, use the method `sort_index()`. This method has the same arguments and default order:

```python
countries_reviewed.sort_index()
```

You can also sort by more than one column at the time:

```python
countries_reviewed.sort_values(by=['country', 'len'])
```

## Data Types and Missing Values

### Dtypes

The data type for a column is known as the **dtype**. You can use the `dtype` property to grab the type of a specific column. 

For example:

```python
wine_reviews.price.dtype
```

In the same way, the `dtypes` property returns the `dtype` of every column in the DataFrame:

```python
wine_reviews.dtypes
```

It is possible to convert a colum of one type into another wherever such a conversion makes sense by using the `astype()` function. For example, we may transform the `points` column from its existing `int64` data type into a `float64` data type:

```python
wine_reviews.points.astype('float64')
```

The *index* of a DataFrame or a Series has its own type, too:

```python
wine_reviews.index.dtype
```

Pandas also support other types of data, such as *categorial data* and *timeseries data*. 

### Missing data

Entries with missing values are given the value `NaN` ('Not a number'). These are always of the `float64` dtype. 

Pandas has some methods to deal with missing data. To select `NaN` ('Not a number') entries one can use the `pd.issnull()` (or its companion `pd.notnull()`)

```python
wine_reviews[pd.isnull(wine_reviews.country)]
```

Replacing mising values is a common operation. Pandas provides the method: `fillna()`. There are many options to use with `fillna()`. For example, replacing `NaN` with 'Unknown':

```python
wine_reviews.region_2.fillna('Unknown')
```

There is also the *backfill strategy*, where we could fill each missing value with the first non-null value that appears sometime after the given record in the database.

We can also replace a non-null value. For that we use the `replace()` method. 

```python
wine_reviews.taster_twitter_handle.replace('@kerinokeefe', '@kerino')
```

## Renaming and Combining

We can change the naming of several properties of a DataFrame. We can also combine data from multiple DataFrames and/or Series.

### Renaming

This is done with the function `rename()`, which lets you change index names and/or column names.  
For example, to change the `points` column in our dataset to `score`, we would do:

```python
wine_reviews.rename(columns={'points': 'score'})
```

We can specify the parameter we want to change: `column` or `index`. The dictionary is the most convenient input format. 

Let's see another example:

```python
wine_reviews.rename(index={0: 'firstEntry', 1: 'secondEntry'})
```

Renaming index values is not that common. For that, `set_index()` is usually more convenient. 

We can change the name of the column index and of the column index with `rename_axis()`. For example:

```python
wine_reviews.rename_axis('wines', axis='rows').rename_axis('fields', axis='columns')
```

### Combining

To combine different DataFrames and Series we can use the methods: `concat()`, `join()`, and `merge()`. Here the focus in on thw first two, as `merge()` can do similar jobs as `join()`.

`concat()` is the simplest one. It requires having the same fields (columns) in the DataFrames or Series you want to combine. Example code:

`canadian_youtube = pd.read_csv("../input/youtube-new/CAvideos.csv")
british_youtube = pd.read_csv("../input/youtube-new/GBvideos.csv")
pd.concat([canadian_youtube, british_youtube])`


`join()` lets you combine different DataFrame object which have an index in common. For example to combine two datasets of videos that happened to be trending on the same day in both Canada and the UK, we could do:


`left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])
left.join(right, lsuffix='_CAN', rsuffix='_UK')`

The `lsuffix` and `rsuffix` parameters are necessary here because the data has the same column names in both datasets. Otherwise we would not need them.
